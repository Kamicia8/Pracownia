```python
import json
import pandas as pd

with open("results1000.json", "r") as f:
    results = json.load(f)

df = pd.DataFrame(results)

denoising_metrics = ["mse_denoised", "rmse_denoised", "mae_denoised", "mape_denoised"]

for metrics in denoising_metrics:

    best_per_function = df.loc[df.groupby("function")[metrics].idxmin()]
    print(f"Najlepsze ustawienia dla każdej funkcji (najniższy {metrics}):")
    print(best_per_function[["function", "T", "beta", "training_loss_function", "epochs", metrics]])

    best_model = df.loc[df[metrics].idxmin()]
    print(f"\nNajlepszy model globalnie (najniższy {metrics}):")
    print(best_model[["function", "T", "beta", "training_loss_function", "epochs", metrics]])

```
